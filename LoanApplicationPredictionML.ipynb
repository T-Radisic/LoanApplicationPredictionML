# Loan Application Success Predictions

## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Data Preparation

#### Importing Dataset

```python
loanApps = pd.read_csv('loan_applications.csv')
features = pd.read_csv('credit_features_subset.csv')
```

#### Sorting and merging datasets

```python
loanApps.sort_values(by='UID', inplace=True)
features.sort_values(by='UID', inplace=True)
```

##### extracting desired columns from datasets

```python
subset1 = loanApps.drop(columns=['ApplicationDate', 'LoanPurpose'])
```

```python
mainDS = pd.merge(subset1, features, on='UID', how='inner')
```

```python
mainDS
```

#### Changing the "EmploymentType" feature to the labels: employed - full time: 0, employed - part time: 1, self employed: 2, retired: 3

```python
mainDS['EmploymentType'] = mainDS['EmploymentType'].replace({'Employed - full time': 0, 'Employed - part time': 1})
mainDS['EmploymentType'] = mainDS['EmploymentType'].replace({'Self employed': 2, 'Retired': 3})
```

```python
mainDS
```

#### Removing any rows that have missing or non-numerical values

```python
mainDS = mainDS.apply(pd.to_numeric, errors='coerce')
mainDS = mainDS.dropna()
```

```python
mainDS
```

#### Splitting dataset into X and Y

##### Success is the Y column. Labels are 0=application passed, 1=application failed

```python
X = mainDS.drop('Success', axis=1)
y = mainDS['Success']
```

```python
X
```

```python
y
```

#### Splitting X and Y for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

## Model Creation

#### Building a random forest classifier and fitting it to our training X and y

```python
from sklearn.ensemble import RandomForestClassifier
RandomForest = RandomForestClassifier(n_estimators=250)
RandomForest.fit(X_train, y_train)
```

#### Testing the classifier

```python
y_pred = RandomForest.predict(X_test)
```

```python
y_pred
```

#### Retrieving Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

## Classifier Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
